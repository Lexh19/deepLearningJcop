```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering, SpectralClustering

from luwiji.cluster import demo
```

# Sample Data

```python
x1, y1 = demo.blob_data()
x2, y2 = demo.moon_data()
x3, y3 = demo.circle_data()
```

# Visualize

```python
x1
```

```python
x = x1
plt.scatter(x[:, 0], x[:, 1], s=10)
plt.axis('equal')
```

# K-means Clustering

```python
kmeans = KMeans(n_clusters=4)

member = kmeans.fit_predict(x)
```

```python
member
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(x[:, 0], x[:, 1], s=10)
plt.scatter(centers[:, 0], centers[:, 1], c='w', edgecolors='k', s=50, linewidths=2)
plt.axis('equal')
```

## K-means weakness
- kesulitan untuk menentukan K
    - Solusi: elbow analysis, sillhoutte score, and others
- hanya untuk data yang bisa dirata ratakan
    - solusi: K-modes, K-medoid
- solusi tidak selalu optimal
- hanya untuk data yang linearly-separable

## contoh: moon dataset

```python
x = x2
kmeans = KMeans(n_clusters=2)
member = kmeans.fit_predict(x)
centers = kmeans.cluster_centers_

plt.figure(figsize=(6, 6))
plt.scatter(x[:, 0], x[:, 1], c=member, s=10)
plt.scatter(centers[:, 0], centers[:, 1], c='w', edgecolors='k', s=50, linewidths=2)
plt.axis('equal')
```

# Menentukan K(n_cluster) tanpa label -> Inertia / elbow analysis

```python
x = x1
score = []
for k in range(2, 25):
    kmeans = KMeans(n_clusters=k)
    member = kmeans.fit_predict(x)
    score.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(2, 25), score, 'r-')
plt.ylabel('inertia')
plt.xlabel('n_clusters')
```

# Menentukan K(n_cluster) tanpa label -> Sillhoutte score

```python
from sklearn.metrics import silhouette_score

x = x1
score = []
for k in range(2, 25):
    kmeans = KMeans(n_clusters=k)
    member = kmeans.fit_predict(x)

    siluet = silhouette_score(x, member, metric='euclidean')
    score.append(siluet)

plt.figure(figsize=(10, 5))
plt.plot(range(2, 25), score, 'r-')
plt.ylabel('silhoutte')
plt.xlabel('n_clusters')
```

# Beyond K-means

## spectral clustering

```python
x = x2
sc = SpectralClustering(n_clusters=2, gamma=100)
labels = sc.fit_predict(x)

plt.figure(figsize=(6, 6))
plt.scatter(x[:, 0], x[:, 1], c=labels, s=10)
plt.axis('equal')
```

masalah terbesarnya adalah kita tidak tau clusternya sudah benar atau belom

## DBSCAN

```python
x = x2
sc = DBSCAN(eps=0.2)
labels = sc.fit_predict(x)

plt.figure(figsize=(6, 6))
plt.scatter(x[:, 0], x[:, 1], c=labels, s=10)
plt.axis('equal')
```

## AHC

```python
x = x2
ahc = AgglomerativeClustering(n_clusters=3, linkage='single')
labels = ahc.fit_predict(x)

plt.figure(figsize=(6, 6))
plt.scatter(x[:, 0], x[:, 1], c=labels, s=10)
plt.axis('equal')
```

# Scipy

```python
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram

x, y = demo.blob_data(30)
z = linkage(x)
```

## get membership

```python
labels = fcluster(z, 4, criterion='maxclust') # berdasarkan maksimal cluster
labels
```

```python
labels = fcluster(z, 5, criterion='distance') # berdasarkan jarak
labels
```

## visualize dendogram

```python
plt.figure(figsize=(15, 5))
dendrogram(z)
```
